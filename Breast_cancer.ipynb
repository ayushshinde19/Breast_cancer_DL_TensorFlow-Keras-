```python
x = bc.drop(columns=["Unnamed: 32", "id", "diagnosis"]) 
y = bc["diagnosis"].map({"M": 1, "B": 0})
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Building Neural Network

```python
import tensorflow as tf
```

```python
tf.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python
#set up the layers of neural network
model = keras.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=(30,)), #input layer
    keras.layers.Dense(20, activation="relu"), #hidden layer
    keras.layers.Dense(10, activation="relu"), #hidden layer
    keras.layers.Dense(1, activation="sigmoid") #output layer
])      
```

```python
#compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
#train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.1, verbose=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend()
plt.show()
plt.plot(history.history["loss"], label="loss")
```

```python
loss = model.evaluate(x_test, y_test, verbose=0) 
```

```python
print("Test loss:", loss[0])
print("Test accuracy:", loss[1])
```

```python
y_pred = model.predict(x_test)
y_pred = np.where(y_pred > 0.5, 1, 0) #convert probabilities to binary values
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Confusion Matrix:")
print(cm)
```
